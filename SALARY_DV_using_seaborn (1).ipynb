**DATASET:**
       
   https://www.kaggle.com/datasets/amirmahdiabbootalebi/salary-by-job-title-and-country

```python
#Import the library:
import pandas as pd
df=pd.read_csv("Salary.csv")
df
```

```python
#import the required module:
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#SCATTER PLOT
sns.scatterplot(x='Age',y='Salary',data=df)
plt.title('SCATTER PLOT')
plt.show()
```

**INTERPRETATION**:

     From the age 40-50 have the highest salary.

```python
#LINE PLOT
sns.lineplot(x='Years of Experience',y='Salary',data=df)
plt.title('LINE PLOT')
plt.show()
```

**INTERPRETATION**:

    Only the employees in the experience of 20-25 years have their salary around 2,25,000 rupees.

```python
#BAR PLOT
sns.barplot(x='Gender',y='Salary',data=df)
plt.title('BARPLOT')
plt.show()
```

**INTERPRETATION**:

    Male employees earn more than Female employees.

```python
#HISTOGRAM
sns.histplot(x='Years of Experience',data=df)
plt.title('HISTOGRAM')
plt.show()
```

**INTERPRETATION**:
      
      The employee whose experience from 2-5 years has the maximum count of 600.






```python
#KDE PLOT
sns.kdeplot(x='Age',data=df)
plt.title('KDEPLOT')
plt.show()
```

**INTERPRETATION**:
      
      Maximum number of employees present at the Age of 20-30.

```python
#BOX PLOT
sns.boxplot(x='Gender',data=df)
plt.title('BOXPLOT')
plt.show()
```

**INTERPRETATION**:

      Equality of Male and Female are same.

```python
#VIOLIN PLOT
sns.violinplot(x='Country',data=df)
plt.title('VIOLINPLOT')
plt.show()
```

**INTERPRETATION**:
     
     The average countries in the dataset are USA,CANADA and CHINA. bold text

```python
#COUNT PLOT
sns.countplot(x='Race',data=df)
plt.title('COUNTPLOT')
plt.show()
```

**INTERPRETATION**:

      The presence of white people employees are more than others.

```python
#PAIR PLOT
sns.pairplot(data=df)
plt.title('PAIRPLOT')
plt.show()
```

**INTERPRETATION**:

     Here we can able to see the various camparison of charts using the pairplot function.
